```python
import pandas as pd
df=pd.read_csv('/content/student-mat.csv')
df.head()
```

```python
df.info()
```

```python
df['G_AVG']=(df['G1']+df['G2']+df['G3'])/3
df.head()
```

```python
df.drop(['romantic','nursery','reason','famsize','G1','G2','G3','Dalc','Walc'],axis=1,inplace=True)
df.head()
```

```python
df['Mjob'].value_counts()
```

```python
df['Mjob']=df['Mjob'].replace({'other':0,'services':1,'at_home':2,'teacher':3,'health':4})
df['Mjob'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

features = ['Fjob','school','address','sex','guardian','schoolsup','famsup','paid','activities','higher','internet','famrel','freetime','goout','health','absences']
for feature in features:
    df[feature] = le.fit_transform(df[feature])
```

```python
le=LabelEncoder()
df['Pstatus']=le.fit_transform(df['Pstatus'])
df['Pstatus']=le.fit_transform(df['Pstatus'])
df.head()
```

```python
y=df['G_AVG']
x=df.drop('G_AVG',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split( x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,lr.predict(x_test))
```
