**Accuracy**----> definition:

the proposion of correct prdiction (both positive and negative)out of all total predictions.

-Equation:
Accuracy =(TP+TN)/(FP+FN+TP+TN)

TP=TRUE POSITIVE

TN=TRUE NEGATIVE

FN=FALSE NEGATIVE

FP=FALSE POSITIVE

**PRECISION**---also called positive prediction Value

Definition:

The propotion of true prositive prediction out of all positive prediciton

PRECISION=TP/(TP+FP)

**RECALL**also called sensitivity or TRUE Positive Rate

Definition:


the porportion of actual positive that are correctly predicted

EQ= Recall=TP/(TP+FN)

**F1 SCORE**

the harmonic mean of precision and recall,balaning the two metrices

EQ= F1 SCORE=2*(PRECISION*RECALL)/(PRECISION*RECALL)



**KEY INFERENCES**:

-high precision mean fewer false positive

-high recall mean fewer false negative

-Accuracy can be misleading when classes are imbalanced

-f1 score provides a balanved measure when precision are recall are equal important
