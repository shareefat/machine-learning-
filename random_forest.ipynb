**Random Forest Alogoritham Intution**:-


it can be divided into 2 stages

in first stage we randomly select "k" feature out of total m Feature and build random forest.in the first stage we proceed as follow:

1. randomly select k feature from total of m feature k < m
2. Among the k feature calculate the node d using the best split point
3. split the node into daughter nodes using the best slipt
4. repeat 1 to 3 steps until 1 number of nodes has been reached
5. Build forest by repeating step 1 to 4 for n number of times to create n number of trees


in second stage , we make prdictions using trained random forest algoritham

1. we take the test features and use the rules os each randomly created decision tree to predict the outcome and store the predicted outcome
2. then we calculate the votes for each predicted target
3. finally we consider the high voted predicted target as the final the random forest algoritham
